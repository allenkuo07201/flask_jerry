```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://tw.stock.yahoo.com/'
```

```python
resp = requests.get(url)
resp
```

```python
resp.text
```

### 使用BeautifulSoup定位

```python
soup = BeautifulSoup(resp.text,'lxml')
soup
```

```python
soup.find('nav',class_ = "indexChartNav")
```

```python
lis = soup.find('nav',class_ = "indexChartNav").find_all('li')
lis
```

```python
li = lis[0]
li
```

```python
li.find_all('span')
```

```python
for span in li.find_all('span'):
    print(span.text.strip())
```

```python
for li in lis:
    for span in li.find_all('span')[:2]:
        print(span.text.strip())
    
```

```python
stocks = []
for li in lis:
    #取前2個資訊
    spans = li.find_all('span')[:2]
    #組合成字典
    data = {"分類":spans[0].text.strip(), 
            "指數":spans[1].text.strip()}
    #加入外部串列
    stocks.append(data)
print(stocks)
```
